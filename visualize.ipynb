```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np

# Load bands first
with rasterio.open('BAND21.tif') as src:
    green = src.read(1)

with rasterio.open('BAND31.tif') as src:
    red = src.read(1)

with rasterio.open('BAND41.tif') as src:
    nir = src.read(1)

# Optional: Normalize for better visualization
green_vis = green / (np.max(green) + 1e-10)
red_vis = red / (np.max(red) + 1e-10)
nir_vis = nir / (np.max(nir) + 1e-10)

# Plot
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

bands = [(green_vis, 'Green (BAND21)'), 
         (red_vis, 'Red (BAND31)'), 
         (nir_vis, 'NIR (BAND41)')]

for i, (band, name) in enumerate(bands):
    axes[i].imshow(band, cmap='gray')
    axes[i].set_title(name)
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```
